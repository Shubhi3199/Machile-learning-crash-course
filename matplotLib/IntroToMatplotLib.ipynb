```python
heartDisease = pd.read_csv('../data/13.1 heart-disease.csv.csv')
```

```python
heartDisease.head()
```

```python
heartDisease['age'].plot.hist();
```

```python
heartDisease.plot.hist(figsize = (15, 25), subplots = True);
```

```python
heartDisease = heartDisease[heartDisease['age'] > 50]
```

```python
# The pyplot way (use this when plotting simpler graphs)
heartDisease.plot(x = 'age',
                 y = 'chol', 
                 c = 'target',
                 kind = 'scatter',
                 xlim = (45, 100));
```

```python
# the object Oriented waywith a mix of pyplot (use this when plotting complex graphs)
fig, ax = plt.subplots()
heartDisease.plot(x = 'age',
                      y = 'chol',
                      c = 'target',
                      kind = 'scatter', 
                      ax = ax);
ax.set_xlim((45, 100))
```

# IMPORTANT

```python
# the object oriented method from scratch (highly customizable recommended)
fig, (ax1, ax2) = plt.subplots(nrows=2, 
                              ncols = 1,
                              figsize = (10, 12),
                              sharex = True)

# plot the data
scatter = ax1.scatter(x = heartDisease['age'],
                    y = heartDisease['chol'],
                    c = heartDisease['target'])

scatter2 = ax2.scatter(x = heartDisease['age'],
                       y = heartDisease['thalach'],
                       c = heartDisease['target'])

# customize the plot
ax1.set(
      ylabel = 'Cholestrol',
      title = 'Heart Diseases and cholestrol levels')

ax2.set(title = 'Heart Disease and Max Heart Rate',
       xlabel = 'age',
       ylabel = 'Max Heart Rate')

# Add a legend
ax1.legend(*scatter.legend_elements(), title = 'target');
ax2.legend(*scatter2.legend_elements(), title = 'target')

# Add a horizontal line
ax1.axhline(heartDisease['chol'].mean(), linestyle = '--');
ax2.axhline(heartDisease['thalach'].mean(), linestyle = '--');

# Add a Titile to the entire Figure
fig.suptitle('Heart Disease Analysis', ha = 'center', fontsize = 26);
```
