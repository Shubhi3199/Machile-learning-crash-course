```python
# Improve the model (Important)
# Trying different amount of n_estimators
np.random.seed(69)
for i in range(10, 110, 10):
    print(f'trying the model with {i} estimators..')
    clf = RandomForestClassifier(n_estimators = i).fit(xTrain, yTrain)
    print(f'MOdel accuracy on the Test Split: {clf.score(xTest, yTest) * 100:.3f}%')
    print('')
```
